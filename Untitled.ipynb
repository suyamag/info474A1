```python
plt.scatter(new['Bacteria '], np.full(16, 0), s=new['Penicilin']*5, alpha=0.75)
plt.scatter(new['Bacteria '], np.full(16, 1), s=new['Neomycin']*5, alpha=0.75)
plt.scatter(new['Bacteria '], np.full(16, 2), s=new['Streptomycin ']*5, alpha=0.75)
plt.xticks(rotation=90)
plt.yticks(np.arange(3), ('Penicilin', 'Neomycin', 'Streptomycin'))
plt.rcParams['figure.figsize']=(15,5)
plt.show()
```

```python
plt.scatter(negative['Bacteria '], np.full(9, 0), s=negative['Penicilin']*5, alpha=0.75, color='red')
plt.scatter(negative['Bacteria '], np.full(9, 1), s=negative['Neomycin']*5, alpha=0.75, color='red')
plt.scatter(negative['Bacteria '], np.full(9, 2), s=negative['Streptomycin ']*5, alpha=0.75, color='red')
plt.scatter(positive['Bacteria '], np.full(7, 0), s=positive['Penicilin']*5, alpha=0.75, color='blue')
plt.scatter(positive['Bacteria '], np.full(7, 1), s=positive['Neomycin']*5, alpha=0.75, color='blue')
plt.scatter(positive['Bacteria '], np.full(7, 2), s=positive['Streptomycin ']*5, alpha=0.75, color='blue')
plt.xticks(rotation=90)
plt.xlabel('Bacteria')
plt.yticks(np.arange(3), ('Penicilin', 'Neomycin', 'Streptomycin'))
plt.ylabel('Antibiotic')
plt.title('MIC for antibiotics on different bacteria')
plt.legend()
plt.rcParams['figure.figsize']=(15,5)

red_patch = mpatches.Patch(color='red', label='Negative Gram Straining')
blue_patch = mpatches.Patch(color='blue', label='Positive Gram Straining')
legend1 = plt.legend(handles=[red_patch, blue_patch], loc='center left', bbox_to_anchor=(1, 0.3))

for area in [10, 100, 800]:
    plt.scatter([], [], c='k', alpha=0.3, s=area*5, label=str(area))
plt.legend(scatterpoints=1, labelspacing=1, title='MIC', loc='center left', bbox_to_anchor=(1, 0.7), prop={'size': 20})
plt.gca().add_artist(legend1)

plt.show()
```
